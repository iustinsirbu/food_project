```python
LABEL = 'INGREDIENT'
from tqdm import tqdm

def process_text(text):
    doc = nlp(str(text)) # handle NaN text
    return ' '.join([token.lemma_ for token in doc])

def df_to_spacy_format(df, show_word=False):
    dataset = []
    for index, row in tqdm(df.iterrows(), total=df.shape[0]):
        text = process_text(row.Directions)
        entities = []
        ingredients = set(row.Ingredients.split(','))
        for ingredient in ingredients:
#             start = text.find(ingredient)
            starts = [m.start() for m in re.finditer(f'{ingredient} ', text)]
            se = [(start, start+len(ingredient)) for start in starts]
            se = sorted(se)
            for start, end in se:
                if show_word:
                    entities.append((start, end, LABEL, ingredient))
                else:
                    entities.append((start, end, LABEL))
                
        entities = sorted(entities)
        last_end = -1
        final_entities = []
        for entity in entities:
            if entity[0] > last_end:
                last_end = entity[1]
                final_entities.append(entity)
        
        entry = (text, {'entities': final_entities})
        dataset.append(entry)

    return dataset
```

```python
from spacy.training import Example
from spacy.util import minibatch, compounding
import random
from tqdm import tqdm


with nlp_model.disable_pipes(*other_pipes):
    sizes = compounding(4.0, 8.0, 1.001)  
    for itn in tqdm(range(10)):
        random.shuffle(train_data)
        batches = minibatch(train_data, size=sizes)
        losses = {}
        
        for batch in batches:
            examples = []
            for text, annotation in batch:
                doc = nlp_model.make_doc(text)
                example = Example.from_dict(doc, annotation)
                examples.append(example)
            nlp_model.update(examples, sgd=optimizer, drop=0.35, losses=losses)
            print("Losses", losses)
```

## Score

```python
import pandas as pd

def compute_metrics(y_true, y_pred):
    y_true = set(y_true)
    y_pred = set(y_pred)
    tp = len(y_true & y_pred)
    
    precision = tp / len(y_pred) if len(y_pred) else 0
    recall = tp / len(y_true) if len(y_true) else 0
    f1 = 2 * tp / (len(y_pred) + len(y_true)) if len(y_pred) + len(y_true) else 0
    
    return precision, recall, f1

def score(model, data):
    metrics = []
    
    for entry in data:
        text, info = entry
        doc = model(text)
        y_pred = [(ent.text, ent.label_) for ent in doc.ents]
        y_true = [(text[ent[0]:ent[1]], ent[2]) for ent in info['entities']]
        p, r, f1 = compute_metrics(y_true, y_pred)
        metrics.append((p, r, f1))
    return pd.DataFrame(metrics, columns=['Precision', 'Recall', 'F1'])
```

```python
test_data = df_to_spacy_format(df_test)
```

```python
# Before training, expected to be 0 as there is no label INGRED
score(nlp, test_data).mean()
```

```python
# After training for 10 epochs. Plateau was not reached, longer training would propably improve the scores.
m = score(nlp_model, test_data)
m
```
